LIBS

```python
import pandas as pd
import numpy as np
```

DADOS

```python
url = r'https://raw.githubusercontent.com/alura-cursos/alura_classificacao_multilabel/refs/heads/master/dataset/stackoverflow_perguntas.csv'
data = pd.read_csv(url)
data
```

```python
individual_tags = [
    tag.strip().lower() for tag in data['Tags']
]
individual_tags
```

```python
individual_tags_list = [sorted(tags.split(' ')) for tags in individual_tags]
individual_tags_list
```

```python
unique_tags = []
for tags in individual_tags_list:
    for tag in tags:
        if tag not in unique_tags and len(tag)>0:
            unique_tags.append(tag)

unique_tags
```

```python
for tag in unique_tags:
    data[tag] = data['Tags'].str.contains(tag, regex=False).astype(int)

data
```

TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
data['all_tags'] = list(zip(
    data[unique_tags[0]],
    data[unique_tags[1]],
    data[unique_tags[2]],
    data[unique_tags[3]],
    ))
data
```

```python
x = data['Perguntas']
y=data['all_tags']

x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    stratify=y,
    random_state=123,
    test_size=0.2,
)
```

VETORIZAR PERGUNTAS

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#max_df retira palavras que aparecem com uma frequencia maior que a definida

vetorizador = TfidfVectorizer(max_features=1000, max_df=0.7)
```

```python
vetorizador.fit(data['Perguntas'])
```

```python
x_train_vectorized = vetorizador.transform(x_train)
x_test_vectorized = vetorizador.transform(x_test)
print(x_train_vectorized.shape)
print(x_test_vectorized.shape)
```

RELEVANCIA BINARIA

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
regressao_logstica = LogisticRegression(solver='lbfgs')
classificador = OneVsRestClassifier(regressao_logstica)
classificador
```

ARRAY Y

```python
y_train = np.asarray(list(y_train))
y_test = np.asarray(list(y_test))
```

TREINAR MODELO

```python
classificador.fit(x_train_vectorized, y_train)
```

METRICAS

EXACT MATCH

```python
result = classificador.score(x_test_vectorized, y_test)
result
```

HAMMING LOSS\
DISTANCIA EM RELACAO AO TOTAL E NAO EM RELACAO A LINHA COMPLETA

```python
from sklearn.metrics import hamming_loss
```

```python
predict = classificador.predict(x_test_vectorized)
result = hamming_loss(y_test, predict)
result
```

CLASSIFICACAO EM CADEIA

```python
from skmultilearn.problem_transform import ClassifierChain
```

```python
chain_classifier = ClassifierChain(regressao_logstica)
```

```python
chain_classifier.fit(x_train_vectorized, y_train)
```

```python
result = chain_classifier.score(x_test_vectorized, y_test)
result
```

```python
predict = chain_classifier.predict(x_test_vectorized)
result = hamming_loss(y_test, predict)
result
```

RELEVANCIA BVINARIA MULTILEARN

```python
from skmultilearn.problem_transform import BinaryRelevance
```

```python
binary_relevance = BinaryRelevance(regressao_logstica)
```

```python
binary_relevance.fit(x_train_vectorized, y_train)
```

```python
result = binary_relevance.score(x_test_vectorized, y_test)
result
```

```python
predict = binary_relevance.predict(x_test_vectorized)
result = hamming_loss(y_test, predict)
result
```

```python
def get_tags(pergunta):
    pergunta = vetorizador.transform([pergunta])
    tags = chain_classifier.predict(pergunta)
    tags = tags.toarray()
    question_tags = []
    for i,tag in enumerate(tags[0]):
        if tag>0:
            question_tags.append(unique_tags[i])
    return question_tags
```
