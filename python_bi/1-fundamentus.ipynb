```python
# Lendo um papel específico
weg = fundamentus.get_papel("WEGE3")
weg
```

```python
todas_acoes = fundamentus.get_resultado_raw().reset_index()
todas_acoes
```

```python
# Definindo a carteira de ações
carteira_fund = ["ABEV3", "B3SA3", "ELET3", "GGBR4", "ITSA4",
                 "PETR4", "RENT3", "SUZB3", "VALE3", "WEGE3"]
```

```python
lista_dfs = []
for ticker in carteira_fund:
    try:
        dados = fundamentus.get_papel(ticker)
        lista_dfs.append(dados)
    except Exception as e:
        print(f"Erro ao coletar dados para {ticker}: {e}")

dados_acoes = pd.concat(lista_dfs, ignore_index=True)
print(dados_acoes)
```

```python
carteira = todas_acoes[todas_acoes['papel'].isin(carteira_fund)]
carteira
```

```python
indicadores_carteira = carteira[['papel','P/L', 'Div.Yield','P/VP','ROE']].reset_index(drop=True)
indicadores_carteira.rename(columns={'papel':'Papel','Div.Yield':'DY'}, inplace= True)
indicadores_carteira
```

```python
# Concatenando os dfs em um só com as infos e indicadores
indicadores = pd.merge(dados_acoes, indicadores_carteira, on="Papel")
indicadores.columns
```

```python
indicadores.to_csv('meus_dados.csv')
```
