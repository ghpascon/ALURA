DATA

```python
import pandas as pd
```

```python
url1 = 'https://raw.githubusercontent.com/alura-cursos/combina-classificadores/main/dados/train.csv'
url2 = 'https://raw.githubusercontent.com/alura-cursos/combina-classificadores/main/dados/test.csv'
```

```python
treino = pd.read_csv(url1)
treino
```

```python
teste = pd.read_csv(url2)
teste
```

TRATAR DADOS

```python
treino.info()
```

```python
treino.isna().sum()
```

```python
teste.isna().sum()
```

```python
def prepara(df):
    df.drop(['Unnamed: 0', 'id'],axis=1,inplace=True)
    df.dropna(inplace=True)
    colunas_categoricas = ["Gender", "Customer Type", "Type of Travel", "Class"]
    df = pd.get_dummies(df, columns=colunas_categoricas)
    X = df.drop('satisfaction', axis=1)
    y = df['satisfaction']
    return X,y
```

```python
x_train, y_train = prepara(treino)
x_test, y_test = prepara(teste)
```

MODELO ARVORE DE DECISAO

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
```

```python
model_1 = DecisionTreeClassifier(
    random_state=563
)
```

```python
validacao = cross_validate(
    model_1,
    x_train,y_train,
    cv=5
)
validacao['test_score'].mean()
```

```python
model_1.fit(x_train, y_train)
model_1.score(x_test, y_test)
```

LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_2 = LogisticRegression(
    random_state=563,
    max_iter=5000
)
```

PIPELINE

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipeline = Pipeline(
    [
        ('Scaler', StandardScaler()),
        ('model', model_2)
    ]
)
```

```python
validacao = cross_validate(
    pipeline,
    x_train,y_train,
    cv=5
)
validacao['test_score'].mean()
```

```python
model_2.fit(x_train, y_train)
model_2.score(x_test, y_test)
```

GAUSSIAN MB

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
model_3 = GaussianNB()
```

```python
modelos = [
    'Arvores',
    'Logistica',
    'Bayes'
]
```

PIPELINE

```python
pipelines = []

for modelo, nome in zip ([model_1,model_2, model_3], modelos):
    pipeline = Pipeline(
        [
            ('scaler', StandardScaler()),
            ('model', modelo)
        ]
    )

    validacao = cross_validate(
        pipeline,
        x_train,y_train,
        cv=5
    )
    print(nome)
    print(validacao['test_score'].mean())

    pipelines.append((nome,pipeline))
```

```python
"""
voting:
hard-> Maior numero de votos
soft-> Define pesos para os modelos
"""

votacao = VotingClassifier(
    estimators=[
        (pipelines[0]),
        (pipelines[1]),
        (pipelines[2]),
    ],
    voting='soft',
)
```

```python
votacao
```

```python
validacao = cross_validate(
    votacao,
    x_train,y_train,
    cv=5
)
validacao['test_score'].mean()
```

TUNING

```python
parametros = {
    'voting':['hard','soft'],
    'weights':[(1,1,1), (2,1,1), (1,2,1), (1,1,2)],
}
```

```python
gridsearch = GridSearchCV(
    votacao, 
    parametros,
    n_jobs=-1
)
gridsearch.fit(
    x_train,y_train
)
```

VALIDACAO CRUZADA DOS MODELOS COMPLEXOS

```python
from mlxtend.classifier import StackingCVClassifier
```

```python
base_models = [
    extra_trees_classifier,
    bagging_classifier,
    model_1
]
```

```python
stacking_cv_classifier = StackingCVClassifier(
    classifiers=base_models,
    meta_classifier=modelo_meta
)
```

```python
mapeamento = {
    'neutral or dissatisfied':0,
    'satisfied':1
}
y_train=y_train.replace(mapeamento)
y_test=y_test.replace(mapeamento)
```

```python
stacking_cv_classifier.fit(x_train, y_train)
```

```python
stacking_cv_classifier.score(x_test,y_test)
```

```python
from mlxtend.evaluate import scoring,confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
```

```python
matriz_confusao = confusion_matrix(
    y_test,
    stacking_cv_classifier.predict(x_test)
) 
```

```python
plot_confusion_matrix(conf_mat=matriz_confusao)
```

```python
metrics = ['accuracy', 'error', 'precision', 'recall', 'f1']
for metric in metrics:
    print(f'Metric: {metric} -> {scoring(y_target=y_test,y_predicted=stacking_cv_classifier.predict(x_test),metric=metric)}')
```
