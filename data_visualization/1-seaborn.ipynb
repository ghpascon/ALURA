DATAFRAME

```python
import pandas as pd
```

```python
df = pd.read_csv('data/canadian_immegration_data.csv')
df
```

```python
df.set_index('Country', inplace=True)
df
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
brasil = df.loc['Brazil', years]
brasil
```

```python
brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes':brasil.values.tolist()
}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

SEABORN

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(
    x='Country',
    y='Total',
    data=top_10
)
```

```python
ax = sns.barplot(
    data=top_10,
    x='Total',
    y='Country',
    orient='h'
)
ax.set(
    title = 'TOP 10 PAISES IMIGRAÇÃO',
    xlabel = 'Imigrantes',
    ylabel = ''
)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax= sns.barplot(
    data=top_10,
    y='Country',
    x='Total',
    orient='h'
)

ax.set(
    title = 'TOP 10 PAISES IMIGRAÇÃO',
    xlabel = 'Imigrantes',
    ylabel = ''
)

plt.show()
```

```python
def gerar_grafico_paleta(paleta):
    fig, ax = plt.subplots(figsize = (8,4))
    ax = sns.barplot(
        data = top_10,
        y = 'Country',
        x = 'Total',
        hue='Total',
        orient = 'h',
        palette = paleta,
        legend=False
    )
    ax.set(
        title = 'TOP 10 PAISES IMIGRAÇÃO',
        xlabel = 'Imigrantes',
        ylabel = '',
    )

    plt.show()
```

```python
gerar_grafico_paleta('Blues')
gerar_grafico_paleta('rocket_r')
gerar_grafico_paleta('coolwarm')
gerar_grafico_paleta('tab10')
```

```python
# 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10_r')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10_r')
```

```python
sns.set_theme(style='darkgrid')
gerar_grafico_paleta('tab10_r')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10_r')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10_r')
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(
    data = top_10,
    y = 'Country',
    x = 'Total',
    hue='Total',
    orient = 'h',
    palette = 'tab10_r',
    legend=False
)

ax.set(
    title = 'TOP 10 PAISES IMIGRAÇÃO',
    xlabel = 'Imigrantes',
    ylabel = ''
)

sns.despine()

plt.show()
```
