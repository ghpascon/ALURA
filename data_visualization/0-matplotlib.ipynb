DATAFRAME

```python
import pandas as pd
```

```python
df = pd.read_csv('data/canadian_immegration_data.csv')
df
```

DEFINIR O PAIS COMO INDEX

```python
df.set_index('Country', inplace=True)
df
```

```python
years = list(map(str, range(1980, 2014)))
years
```

SELECIONANDO O BRASIL

```python
brasil = df.loc['Brazil', years]
brasil
```

```python
brasil_dict = {
    'ano': brasil.index.tolist(),
    'imigrantes':brasil.values.tolist()
}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

GRAFICO DE LINHA

```python
plt.figure(figsize=(8,4))
plt.plot(
    dados_brasil['ano'],
    dados_brasil['imigrantes'],
    color='blue',
    marker='o',
    markersize=5
)

plt.title('Imigração Brasil-Canada')
plt.xlabel('Anos')
plt.ylabel('Imigrantes')

plt.xticks([year for year in years if int(year)%5 == 0])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(
    dados_brasil['ano'],
    dados_brasil['imigrantes'],
    color='blue',
    marker='o',
    markersize=5
)
ax.set_title('Imigração Brasil-Canada')
ax.set_xlabel('Ano')
ax.set_ylabel('Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20,7))

axs[0].plot(
    dados_brasil['ano'],
    dados_brasil['imigrantes'],
    color='blue',
    marker='o',
    markersize=5
)
axs[0].set_title('Imigração Brasil-Canadá]')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot Imigração Brasil-Canada')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração America do Sul - Canadá')

axs[0,0].plot(df.loc['Brazil', years])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Argentina', years])
axs[0,1].set_title('Argentina')

axs[1,0].plot(df.loc['Peru', years])
axs[1,0].set_title('Peru')

axs[1,1].plot(df.loc['Colombia', years])
axs[1,1].set_title('Colombia')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')
    ax.set_ylim(0, 7000)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(
    dados_brasil['ano'],
    dados_brasil['imigrantes'],
    color='#ff00ff',
    marker='o',
    markersize=5,
    lw=3
)
ax.set_title('Imigração Brasil-Canada', fontsize=22)

ax.set_xlabel('Ano', fontsize=12)
ax.xaxis.set_tick_params(labelsize=10)

ax.set_ylabel('Imigrantes', fontsize=12)
ax.yaxis.set_tick_params(labelsize=10)

ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.grid(linestyle='--')
plt.show()
```

AMERICA DO SUL

```python
america_sul = df[df['Region'] == 'South America']
america_sul = america_sul.sort_values('Total', ascending=True)
america_sul
```

GRAFICO DE BARRAS

```python
fix, ax = plt.subplots(figsize=(12,5))

ax.bar(
    america_sul.index,
    america_sul['Total'],
    color = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato'],
)
ax.set_title('Imigração América do Sul-Canadá', loc='left', fontsize=18)
ax.set_xlabel('')
ax.set_ylabel('Imigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
cores = []
for pais in america_sul.index:
    if pais == 'Brazil':
        cores.append('g')
    else:
        cores.append('#cccccc')

fix, ax = plt.subplots(figsize=(12,5))

ax.barh(
    america_sul.index,
    america_sul['Total'],
    color = cores,
)
ax.set_title('Imigração América do Sul-Canadá', loc='left', fontsize=18)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,v in enumerate(america_sul['Total']):
    ax.text(
        v + 50,
        i,
        str(v),
        color='black',
        fontsize=10,
        ha = 'left',
        va = 'center'
    )
    
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)

ax.tick_params(axis='both', which='both', length=0)
plt.show()
```
