```python
import mysql.connector
from mysql.connector import Error
import json
```

DB DATA

```python
with open("config/db.json") as db_file:
    db_config = json.load(db_file)
```

CONNECTION

```python
def get_db_connection():
    return mysql.connector.connect(
            host=db_config['host'],
            user=db_config['user'],
            password=db_config['password'],
            database=db_config['database']
        )
```

```python
def execute_query(query, params=None):
    try:
        # Create connection to the database
        conn = get_db_connection()

        with conn.cursor() as cursor:
            # Use parameterized query if params are provided
            if params:
                cursor.execute(query, params)
            else:
                cursor.execute(query)

            # For SELECT queries, fetch the results
            if cursor.description:  # This indicates a SELECT query
                return cursor.fetchall()

            # For data-modifying queries, commit the transaction
            if query.strip().lower().startswith(('insert', 'update', 'delete')):
                conn.commit()
                return cursor.rowcount  # Return number of affected rows
            
            return None

    except Error as err:
        print(f"Error executing query: {err}")
        raise  # Re-raise error for further handling if necessary

    finally:
        if conn.is_connected():
            conn.close()
```

TABELAS

```python
tabelas = [
    'alugueis',
    'avaliacoes',
    'clientes',
    'enderecos',
    'hospedagens',
    'proprietarios'
]
```

SELECIONAR TODOS OS DADOS DA TABELA

```python
query = f'SELECT * FROM {tabelas[1]}'
result = execute_query(query)
result
```

WRITE NA TABELA

```python
table = 'enderecos'
data = ['225566', 'Trecho de Moraes', '5', 'Baleia', 'Teixeira', 'RR', '62030-547']
data_count = r'%s,'*len(data)
data_count=data_count[:-1]

# Parameterized query
query = f"INSERT INTO {table} VALUES ({data_count})"
params = data

# Execute the query with parameters
result = execute_query(query, params)
if result>0:
    print('Sucesso')
```

SELECIONANDO COLUNAS ESPECIFICAS

```python
colunas = ['data_inicio', 'data_fim']
colunas = ','.join(colunas)
query = f'SELECT {colunas} FROM {tabelas[0]}'
result = execute_query(query)
print(result)
```

TIPOS

```python
query = f"""
SELECT * FROM hospedagens 
WHERE tipo = 'hotel' AND
ativo = 1
"""
result = execute_query(query)
result
```

AVALIACOES

```python
query = f"""
SELECT * FROM avaliacoes 
WHERE nota >= 4
"""
result = execute_query(query)
result
```

PRECOS

```python
query = f"""
SELECT cliente_id, AVG(preco_total) as ticket_medio
FROM alugueis 
GROUP BY cliente_id
"""
result = execute_query(query)
result
```

DIAS

```python
#ORDER BY -> ASC, DESC

query = f"""
SELECT cliente_id, AVG(DATEDIFF(data_fim, data_inicio)) as dias
FROM alugueis 
GROUP BY cliente_id
ORDER BY dias DESC
"""
result = execute_query(query)
result
```

-----  JUNTANDO TABELAS  -----

TOP 10 HOSPEDAGENS ATIVAS

```python
query = f"""
SELECT p.nome AS nome_proprietario, COUNT(h.hospedagem_id)
AS total_hospedagens
FROM proprietarios p
JOIN hospedagens h
ON p.proprietario_id = h.proprietario_id
WHERE h.ativo = 1
GROUP BY p.nome
ORDER BY total_hospedagens DESC
LIMIT 20
"""
result = execute_query(query)
result
```

IMOVEIS INATIVOS

```python
query = f"""
SELECT p.nome AS nome_proprietario, COUNT(h.hospedagem_id)
AS total_hospedagens
FROM proprietarios p
JOIN hospedagens h
ON p.proprietario_id = h.proprietario_id
WHERE h.ativo = 0
GROUP BY p.nome
ORDER BY total_hospedagens DESC
LIMIT 20
"""
result = execute_query(query)
result
```

MESES COM MAIOR DEMANDA

```python
query = f"""
SELECT MONTH(data_inicio) AS mes, COUNT(*) as total_alugueis
FROM alugueis
GROUP BY mes
ORDER BY total_alugueis DESC
"""
result = execute_query(query)
result
```

```python
query = f"""
SELECT MONTH(data_inicio) AS mes, COUNT(*) as total_alugueis
FROM alugueis
JOIN hospedagens 
ON alugueis.hospedagem_id = hospedagens.hospedagem_id 
WHERE hospedagens.tipo = 'hotel'
GROUP BY mes
ORDER BY total_alugueis DESC
"""
result = execute_query(query)
result
```
