LIBS

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

ABRIR IMAGEM

```python
img = cv.imread('data/girl.png')
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img, cmap='gray')
```

CLASSIFICADOR

```python
classificador = cv.CascadeClassifier('classificador/haarcascade_frontalface_default.xml')
print(classificador)
```

EXTRAIR REGIAO DE INTERESSE

```python
scaler = 1.3
min_features = 5
faces = classificador.detectMultiScale(gray_img, scaler, min_features)
print(len(faces))
print(faces)
```

```python
face_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
rect_color = (0,255,0)
for x,y,w,h in faces:
    cv.rectangle(face_img, (x,y), (x+w, y+h), rect_color, 10)

plt.imshow(face_img)
    
```
